```python
def temp(x):
    if x == 80:
        return x
    elif x == 100:
        return x
    else:
        return 200
```

```python
df = pd.read_csv('joined.csv').drop('Unnamed: 0', axis=1)
df
```

```python
# fill NaN in item with quantile/proportional values
for num in item_list:
    if num == 3450:
        df[str(3450)] = df[str(3450)].apply(lambda x: temp(x))
        continue
    if num in item_list7+item_list8:
        df[str(num)] = clean_numeric(df[str(num)])
    is_continuous = 0 if df[str(num)].dtype == 'object' else 1
    if len(df[str(num)].unique())<=10:
        is_continuous = 0
    df = fill_missing_mean(df, str(num), is_continuous)
df
```
