```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from dataprep.eda import plot
import math
import random
from utils import *
```

```python
base = pd.read_csv('cohort.csv').drop('Unnamed: 0', axis=1)
base
```

```python
data = base
item_list1 = [646, 50868, 212, 161, 128, 550, 1125, 159]
item_list2 = [1484, 8368, 5815, 8549, 5820,8554,5819,8553]
item_list3 = [834, 581, 8441, 456]
item_list4 = [31, 5817, 8551, 113, 1703, 467, 80, 1337, 674, 432]
item_list5 = [5813, 8547,617, 210, 637, 198, 707]
item_list6 = [704, 479, 54, 32, 547, 154]
item_list7 = [51221,50912,50902,51265,51222,50931,51249,51279,51248]
item_list = item_list1+item_list2+item_list3+item_list4+item_list5+item_list6+item_list7
```

```python
for num in item_list:
    table = pd.read_csv('data/NEW_ITEMS/ITEM_'+str(num)+'.csv')
    table = table.drop_duplicates('HADM_ID')

    try:
        table = table.rename(columns={"VALUE": str(num)}).drop(['Unnamed: 0', 'ICUSTAY_ID', 'SUBJECT_ID', 'ROW_ID', 'ITEMID', 'CHARTTIME', 'VALUENUM', 'VALUEUOM'], axis=1)
    except:
        table = table.rename(columns={"VALUE": str(num)}).drop(['Unnamed: 0', 'SUBJECT_ID', 'ROW_ID', 'ITEMID', 'CHARTTIME', 'VALUENUM', 'VALUEUOM'], axis=1)
    data = data.merge(table, on='HADM_ID', how='left')
data
```

```python
# fill NaN in item with mean/proportional values
for num in item_list:
    if num in item_list7:
        data[str(num)] = clean_numeric(data[str(num)])
    is_continuous = 0 if data[str(num)].dtype == 'object' else 1
    if len(data[str(num)].unique())<=10:
        is_continuous = 0
    data = fill_missing_mean(data, str(num), is_continuous)
data
```

```python
categorical = []
for num in item_list:
    if data[str(num)].dtype == 'object':
        categorical.append(str(num))
categorical
```

```python
onehot = pd.get_dummies(data, columns=categorical)
onehot
```
